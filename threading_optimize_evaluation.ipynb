```python
import pandas as pd
import numpy as np
#%matplotlib tk
import matplotlib.pyplot as plt
from chardet import detect
rawdata = "thread_optimize_data.csv"
with open(rawdata, 'rb') as f: 
    b = f.read()  
det_results=detect(b)
print(det_results)
df_data = pd.read_csv(rawdata, encoding=det_results["encoding"])
print(df_data.dtypes)
print(df_data)
total_files = [500, 1000, 1500]
temp=df_data[df_data["total_files"]== total_files[0] ]
x = temp.iloc[:, 0]
print(x)
YT_KEY = 2
yt = temp.iloc[:,YT_KEY]
yx = temp.iloc[:,YT_KEY+1]
yf = temp.iloc[:,YT_KEY+2]
print(yx)
keys=temp.keys()
print(keys)
print(keys[YT_KEY])
```

```python
import figsetting as fs
print(fs)
CONFIG = fs.CONFIG
print(CONFIG)
plt.rcParams.update(CONFIG)
FONT_LABEL = fs.FONT_LABEL
print(FONT_LABEL)
FONT_TITLE = fs.FONT_TITLE

```

```python
LABEL_X = "Max_workers"
LABEL_Y = "Time"
GRAPH_TITLE = "Total_files = 500"
fig, ax = plt.subplots(1,1, figsize = fs.FIGSIZE)
# Figure
#fig = plt.figure()

# Axes
#ax = fig.subplots()

# 
#ax.plot(x, yt)
ax.plot(x, yt, color = fs.LINE_COLOR, label=keys[YT_KEY])
ax.plot(x, yx, color = fs.LINE_COLOR2, label=keys[YT_KEY+1])
ax.plot(x, yf, color = 'blue', label=keys[YT_KEY+2])
# X
ax.set_xlabel(LABEL_X, **FONT_LABEL)
ax.minorticks_on()
ax.tick_params(direction="in", which="both", length=5, labelsize=16)
ax.grid(which="both", axis="both", color="gray", linestyle="--", linewidth=0.6)
# Y
ax.set_ylabel(LABEL_Y, **FONT_LABEL)
ax.set_title(GRAPH_TITLE, **FONT_TITLE)

ax.legend( loc="upper left")

# 
fig.show()
```
