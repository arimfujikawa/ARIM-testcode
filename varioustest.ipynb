```python
import numpy as np
import time
import os 

NORM_DIRECTORY = '03/fit'
FT_DIRECTORY = '03/fit'
FIT_DIRECTORY = '03/fit'
CHIK_DIRECTORY = '03/fit'
NORM = 'norm'
CHIK = 'chik'
FT = 'FT'
FIT = 'Fit'
FILE_SAFIX = '_results.npz'
HEADER_SAFIX = '_header.npy'
directories = [f'{NORM_DIRECTORY}/{NORM}', f'{CHIK_DIRECTORY}/{CHIK}', f'{FT_DIRECTORY}/{FT}', f'{FIT_DIRECTORY}/{FIT}' ]
prefixes = [NORM, CHIK, FT, FIT]
files = []
headers = []

def save_to_text(directory, loaded_npz, header_list):
    if len(loaded_npz) == len(header_list):
        for l_n, header_l in zip(loaded_npz, header_list):
            np.savetxt(f'{directory}/{l_n}', loaded_npz[l_n], header=header_l, delimiter='\t')
    else:
        print(f'list length is not match between npz ({len(loaded_npz)}) and header_list ({len(header_list)})')
    
start_time = time.time()
for directory, prefix in zip(directories, prefixes):
    files.append(f'{directory}/{prefix}{FILE_SAFIX}')
    headers.append(f'{directory}/{prefix}{HEADER_SAFIX}')
for file, directory, header in zip(files, directories, headers):
    try:
        save_to_text(directory, np.load(file), np.load(header))
        print(f'{file} was finished')
    except FileNotFoundError:
        basename = os.path.basename(file)
        header
        print(f'{basename} or is not found ! Arrange {basename} in {directory}
    
print( time.time()-start_time)

```

```python

import time
import functools
from concurrent.futures import ProcessPoolExecutor

def call_slow_request():
    count = 0
    for i in range(10):
        time.sleep(0.1)
        count += 1
    count+=1
    for i in range(10):
        time.sleep(0.1)
        count += 1
    print(count)
    return 'ok'

def processing_time(func):
    @functools.wraps(func)
    def wrapper(*args, **keywords):
        st = time.time()  # 開始前の時間を記録
        result = func(*args, **keywords)  # 関数を実行
        print(f'time: {time.time() - st} s')  # 開始後の時間と開始前の時間の差を出力
        return result

    return wrapper

@processing_time
def run_concurrent():
    with ProcessPoolExecutor() as executor:
        features = [executor.submit(call_slow_request) for _ in range(3)]
        for feature in features:
            print(feature.result())


run_concurrent()
```

```python
import random
import threading
import time

from line_profiler import LineProfiler


def do_other_stuff(numbers):
    time.sleep(1)
    s = sum(numbers)


def do_stuff(numbers):
    my_thread = threading.Thread(target=do_other_stuff, args=[numbers])
    my_thread.start()
    my_thread.join()


thread_to_lp = {}


def wrapper_in_thread(func):
    def inner(*args, **kwargs):
        lp = thread_to_lp.setdefault(threading.current_thread(), LineProfiler())
        lp.add_function(func)
        lp.enable_by_count()
        try:
            return func(*args, **kwargs)
        finally:
            lp.disable_by_count()

    return inner


do_other_stuff = wrapper_in_thread(do_other_stuff)
do_stuff = wrapper_in_thread(do_stuff)

numbers = [random.randint(1, 100) for i in range(1000)]
do_stuff(numbers)

for lp in thread_to_lp.values():
    lp.print_stats()
```
