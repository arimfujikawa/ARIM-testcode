### 2次元のマトリクスからgnuplotで読めるように変換

```python
import scipy
from scipy.stats import multivariate_normal

start_x = 0
end_x = 10 
size_x = 11
step_x = 1
start_y = 20
end_y = 40
step_y = 2
size_y = 11

mu = (5, 30)
P = (7, 15)

with open('gaussian.dat', 'w', encoding='UTF-8') as fs:
    for i in range(size_x):
        x = start_x + (i * step_x)
        for j in range(size_y):
            y = start_y + (j * step_y)
            pos = (x, y)
            print(f'{x} {y} {multivariate_normal(mu, P).pdf(pos):.4f}')
            fs.write(f'{x} {y} {multivariate_normal(mu, P).pdf(pos):.4f}\n')
        
# with

# print(f'{multivariate_normal(mu, P).pdf(x):.4f}')
```
