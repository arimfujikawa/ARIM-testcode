# 環境変数のパス

### フォントのパス

```python
import sys
import os
if sys.platform == "win32":
    d = os.environ.get("WINDIR")
    print([os.path.join(d, "fonts")])
```

```python
from matplotlib import font_manager
fonts = font_manager.findSystemFonts()
print([[str(font), font_manager.FontProperties(fname=font).get_name()] for font in fonts])
```
