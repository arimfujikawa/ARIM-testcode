```python
# %matplotlib inline

import tkinter as tk
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk)

# ウィンドウをクローズしたときrootを消去する
def destroy_window():
    root.quit()
    root.destroy()
    

# ガウシアンの定義
def gauss(x, amp=1, center=0, sigma=1):
    return amp * np.exp(-(x - center)**2 / (2*sigma**2))


# データの読み込み
data = np.genfromtxt("test001.dat") 
print(f'{data}:{type(data)}')
# x,y軸の定義
x = data[:,0]
y = data[:,1]
print(f'x = {x}\ny = {y}')

# グラフ領域の定義
fig = plt.figure(figsize=(6, 4))
axis = fig.add_subplot(111)
# グラフ作成
# マーカーの中抜き設定
axis.plot(x, y, 'o', markeredgecolor='black', markeredgewidth=0.5, fillstyle='none', label='raw')


    
# rootウィンドウを作成
root = tk.Tk()
root.withdraw()
root.protocol('WM_DELETE_WINDOW', destroy_window)

# キャンバスの設定
# キャンバスへグラフの埋め込み
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.draw()
toolbar = NavigationToolbar2Tk(canvas, root)
canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

toolbar.pack(side=tk.BOTTOM, fill=tk.BOTH)

root.update()
# 再表示
root.deiconify()
root.mainloop()
```

### Tkinterを利用してツールバーを加えた

```python
# %matplotlib inline

import tkinter as tk
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk)

# データの読み込み
data = np.genfromtxt("test001.dat") 
print(f'{data}:{type(data)}')
# x,y軸の定義
x = data[:,0]
y = data[:,1]
print(f'x = {x}\ny = {y}')

# グラフ領域の定義
fig = plt.figure(figsize=(6, 4))
axis = fig.add_subplot(111)
# グラフ作成
# マーカーの中抜き設定
axis.plot(x, y, 'o', markeredgecolor='black', markeredgewidth=0.5, fillstyle='none', label='raw')

# ウィンドウをクローズしたときrootを消去する
def destroy_window():
    root.quit()
    root.destroy()
    
    
# rootウィンドウを作成
root = tk.Tk()
root.withdraw()
root.protocol('WM_DELETE_WINDOW', destroy_window)

# キャンバスの設定
# キャンバスへグラフの埋め込み
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.draw()
toolbar = NavigationToolbar2Tk(canvas, root)
canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

toolbar.pack(side=tk.BOTTOM, fill=tk.BOTH)

root.update()
# 再表示
root.deiconify()
root.mainloop()
```
