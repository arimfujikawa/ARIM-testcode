# jsonファイルから要素を抜き出す

### keyとvalueを表に出力
- keyの階層に合わせてkey1、key2・・・と列を作成
1. jsonのファイルを読み込み
1. pandasのデータフレームを準備して、value列を定義しておく
1. dict形式を再帰関数でデータフレームに要素を入力
1. CSVでデータフレームを出力

```python
import json
import pandas as pd

line = 0
def make_table(data, index=0, dataframe=None):
    global line
    if isinstance(data, dict):
        for key,value in data.items():
            print(f'key,value {line}:{index} = {key}, {value}')
            # 列を追加
            if len(dataframe.index) < index:
                # 列が重複しないか確認
                if len(df.filter(items=[f'key{index}']).columns) == 0:
                    dataframe[f'key{index}'] = ''
            dataframe.loc[line, f'key{index}'] = key         
            # 新たなkeyができるので列を増やす
            make_table(value, index+1, dataframe)
           

    else:
        print(f'value {line}:{index} = {data}')
        dataframe.loc[line, 'value'] = data
        # valueまでの再帰処理が終わったので行を増やす
        line += 1
    
    return 

json_file = 'json_file/metadata-def_all.json'
with open(json_file, 'r', encoding='utf-8') as f:
    dict_object = json.load(f)
print(dict_object)
df = pd.DataFrame()
df['value'] = ''
make_table(dict_object, dataframe=df)
print(df)
# shift-jisにエンコード
df.to_csv('dataframe.csv', mode='w', sep=',', encoding="shift jis")
```

### ARIMメタデータのjpとenのみ抜きだす

```python
import json
import pandas as pd

line = 0
def make_table(data, dataframe=None):
    global line
    if isinstance(data, dict):
        for key,value in data.items():
            print(f'key,value {line} = {key}, {value}')
            if key == 'ja' or key == 'en':
                dataframe.loc[line, f'key'] = key
                print(f'value {line} = {value}')
                dataframe.loc[line, 'value'] = value
                line += 1      
            # 新たなkeyができるので列を増やす
            make_table(value, dataframe)
    else:
        print(f'value {line} = {data}')

    
    return 

json_file = 'json_file/metadata-def_all.json'
with open(json_file, 'r', encoding='utf-8') as f:
    dict_object = json.load(f)
print(dict_object)
df = pd.DataFrame()
df['value'] = ''
df['key'] = ''
make_table(dict_object, dataframe=df)
print(df)
# shift-jisにエンコード
df.to_csv('dataframe.csv', mode='w', sep=',', encoding="shift jis")
```
